```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame ensuring utf-8 encoding
df = pd.read_csv('NALA_full_data.csv', encoding='utf-8')
df.head()
```

```python
df.info()
```

```python
# see rows with values for user_evaluation and user_feedback
df[df['user_evaluation'].notna() | df['user_feedback'].notna()]
```

```python
df.drop(['user_evaluation', 'user_feedback', 'triggered_by'], axis=1, inplace=True)
len(df)
```

```python
# find which rows have empty values for 'text' column
empty_text_rows = df[df['text'].isna()]
print(len(empty_text_rows))
empty_text_rows
```

```python
# drop rows with empty 'text' values
df.dropna(subset=['text'], inplace=True)
len(df)
```

```python
df['sender'].value_counts()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.info()
```

```python
# change to sgt timezone
df['timestamp_sgt_copy'] = df['timestamp'].dt.tz_convert('Asia/Singapore')
df.head()
```

```python
df.info()
```

```python
# drop timestamp_sgt and timestamp column and rename timestamp_sgt_copy to timestamp
df.drop(columns=['timestamp_sgt', 'timestamp'], inplace=True)
df.rename(columns={'timestamp_sgt_copy': 'timestamp_sgt'}, inplace=True)
df.info()
```

```python
# print the first row of context column
print(df['context'][0])
```

```python
# drop context column
df.drop(columns=['context'], inplace=True)
df.info()
```

```python
# srt by email and then by conversation id then within conversation sort by timestamp
df = df.sort_values(by=['email', 'conversation_id', 'timestamp_sgt'])
df.head(20)
```

```python
# function to clean the text column by removing newlines and extra spaces
# decode unicode escape sequences and ensure the text is in utf-8 encoding
def clean_text(text):
    if isinstance(text, str):
        # replace newlines with spaces and strip whitespace
        text = text.replace('\n', ' ').strip()
        # decode unicode escape sequences and ensure the text is in utf-8 encoding
        cleaned_text = bytes(text, 'utf-8').decode('utf-8', 'replace')
        return cleaned_text
    else:
        return text
    
# apply the clean_text function to the text column
df['text'] = df['text'].apply(clean_text)
df.head(20)
```

```python
# save the cleaned dataframe to a new csv file
df.to_csv('NALA_full_data_cleaned.csv', index=False, encoding='utf-8')
```
